```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
events = pd.read_csv('120-years-of-olympic-history-athletes-and-results/athlete_events.csv')
regions = pd.read_csv('120-years-of-olympic-history-athletes-and-results/noc_regions.csv')

events.head()
```

```python
events.isnull().sum()
```

```python
events.describe()
```

```python
attr = ['Sex', 'Age', 'Height', 'Weight', 'Country', 'Year', 'Season', 'Sport', 'Medal']
eventsSelect = events[['Sex', 'Age', 'Height', 'Weight', 'NOC', 'Year', 'Season', 'Sport', 'Medal']]
attr_arr = eventsSelect.values.tolist()
eventsFilter = []
for i in attr_arr:
    if pd.isnull(i[1]) or pd.isnull(i[2]) or pd.isnull(i[3]):
        pass
    else:
        eventsFilter.append(i)

# print(eventsFilter)
eventsFilter2 = [i for i in eventsFilter]
for i in range(len(eventsFilter2)):
    eventsFilter2[i][1] = float(eventsFilter2[i][1])
    eventsFilter2[i][2] = float(eventsFilter2[i][2])
    eventsFilter2[i][3] = float(eventsFilter2[i][3])
    eventsFilter2[i][5] = float(eventsFilter2[i][5])
    if eventsFilter2[i][6] == 'Summer':
        eventsFilter2[i][6] = 0
    elif eventsFilter2[i][6] == 'Winter':
        eventsFilter2[i][6] = 1

    if eventsFilter2[i][0] == 'M':
        eventsFilter2[i][0] = 0
    elif eventsFilter2[i][0] == 'F':
        eventsFilter2[i][0] = 1

    if eventsFilter2[i][8] == 'Gold':
        eventsFilter2[i][8] = 3
    elif eventsFilter2[i][8] == 'Silver':
        eventsFilter2[i][8] = 2
    elif eventsFilter2[i][8] == 'Bronze':
        eventsFilter2[i][8] = 1
    else:
        eventsFilter2[i][8] = 0

country = regions[['NOC']]
country = country.values.tolist()
countryMap = []
for i in range(len(country)):
    countryMap.append([i + 1, country[i]])

# print(countryMap)

sport = events[['Sport']]
# print(sport)
sportList = sport.values.tolist()
print(type(sportList))
sportListCopy = []
sportSet = []
for i in range(len(sportList)):
    sportListCopy.append(sportList[i][0])

# print(sportListCopy)
sportListCopy = sorted(sportListCopy)
for i in sportListCopy:
    if i not in sportSet:
        sportSet.append(i)

sportMap = []
for i in range(len(sportSet)):
    sportMap.append([i + 1, sportSet[i]])
    
# print(sportMap)

for i in range(len(eventsFilter2)):
    for j in range(len(sportMap)):
        if eventsFilter2[i][7] == sportMap[j][1]:
            eventsFilter2[i][7] = sportMap[j][0]

for i in range(len(eventsFilter2)):
    for j in range(len(countryMap)):
        if eventsFilter2[i][4] == countryMap[j][1][0]:
            eventsFilter2[i][4] = float(countryMap[j][0])
            
for i in range(10):
    print(eventsFilter2[i])
df = pd.DataFrame(eventsFilter2, columns=attr)
df['Country'] = df['Country'].astype('float')
df.head()
```

```python
playerAndCounty = events[['ID', 'Name', 'NOC', 'Medal']]
goldPlayers = playerAndCounty[playerAndCounty['Medal'] == 'Gold']
silverPlayers = playerAndCounty[playerAndCounty['Medal'] == 'Silver']
bronzePlayers = playerAndCounty[playerAndCounty['Medal'] == 'Bronze']

goldPlayers = list(goldPlayers["NOC"])
goldPlayersCountry = list(set(goldPlayers))
silverPlayers = list(silverPlayers["NOC"])
silverPlayersCountry = list(set(silverPlayers))
bronzePlayers = list(bronzePlayers["NOC"])
bronzePlayersCountry = list(set(bronzePlayers))
goldCounter = [[0 for j in range(5)] for i in range(len(goldPlayersCountry))]
# print(goldCounter)

for i in range(0, len(goldPlayersCountry)):
    goldCounter[i][0] = goldPlayersCountry[i]
for i in range(0, len(goldPlayers)):
    for j in range(0, len(goldCounter)):
        if goldPlayers[i] == goldCounter[j][0]:
            goldCounter[j][1] += 1

temp = [i[0] for i in goldCounter]
for i in range(0, len(silverPlayersCountry)):
    if silverPlayersCountry[i] not in temp:
        goldCounter.append([silverPlayersCountry[i], 0, 0, 0, 0])
for i in range(0, len(silverPlayers)):
    for j in range(0, len(goldCounter)):
        if silverPlayers[i] == goldCounter[j][0]:
            goldCounter[j][2] += 1

temp = [i[0] for i in goldCounter]

for i in range(0, len(bronzePlayersCountry)):
    if bronzePlayersCountry[i] not in temp:
        goldCounter.append([bronzePlayersCountry[i], 0, 0, 0, 0])
for i in range(0, len(bronzePlayers)):
    for j in range(0, len(goldCounter)):
        if bronzePlayers[i] == goldCounter[j][0]:
            goldCounter[j][3] += 1

for i in range(len(goldCounter)):
    goldCounter[i][4] = goldCounter[i][1] + goldCounter[i][2] + goldCounter[i][3]

goldCounter = sorted(goldCounter, key=lambda x: x[4], reverse=True)
medalCounter = [i for i in goldCounter if i[4] > 100]
# goldCounters = []

from pyecharts import Bar
bar = Bar("medal count",  width=1000, height=600)
attr = [i[0] for i in medalCounter]
vg = [i[1] for i in medalCounter]
vs = [i[2] for i in medalCounter]
vb = [i[3] for i in medalCounter]
vSum = [i[4] for i in medalCounter]
# print(attr)
bar.add("gold", attr, vg, label_color=["blue"], is_stack=True)
bar.add("silver", attr, vs, label_color=["red"], is_stack=True)
bar.add("bronze", attr, vb, label_color=["silver"], is_stack=True)
bar.add("total", attr, vSum, label_color=["gold"], is_stack=False)

bar.render()
```
